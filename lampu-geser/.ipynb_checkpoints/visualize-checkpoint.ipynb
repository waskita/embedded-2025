```python
# tutorial step graph:
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/step_demo.html
import matplotlib.pyplot as plt
import numpy as np
```

```python
# bikin sinyal , output ke csv
```

```python
#jalankan simulator di C, output ke toggle_output.csv
import pandas as pd
df = pd.read_csv('geser.csv')
```

```python
# figure size: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/figure_size_units.html
# https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html
# plt.subplots(figsize=(8, 4))
# multiple subgraph demo: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/multiple_figs_demo.html
plt.figure(figsize=(10,6))
plt.subplot(311)
plt.step(df["k"], df["geser_input_kiri"], where='post',color = 'green', label = "input") 
plt.plot(df["k"], df["geser_input_kiri"], 'o--',color = 'grey', label = "input", alpha=0.3) 
plt.subplot(312)
plt.step(df["k"], df["geser_input_kanan"], where='post',color = 'blue', label = "input") 
plt.plot(df["k"], df["geser_input_kanan"], 'o--',color = 'grey', label = "input", alpha=0.3) 
plt.subplot(313)
plt.step(df["k"], df["geser_state"], where='post',color = 'red', label = "output") 
plt.plot(df["k"], df["geser_state"], 'o--',color = 'red', label = "input", alpha=0.3) 
plt.subplot(414)
plt.step(df["k"], df["geser_state"], where='post',color = 'red', label = "output") 
plt.plot(df["k"], df["geser_state"], 'o--',color = 'red', label = "input", alpha=0.3) 

#plt.step(x, y, where='post', label='post')
plt.savefig("geser.jpg")
plt.show()
```
