```python
from PIL import Image, ImageDraw

def analyze_image(work_input_filename,work_output_filename):
    print(work_input_filename,work_output_filename)
    image = Image.open(work_input_filename)
    
    # Get image dimensions
    width, height = image.size

    output_image  = Image.new(mode="RGB", size=(width, height), color=(255, 255, 255))

    pixels_r=[]
    pixels_g=[]
    pixels_b=[]
    
    delta_r=[]
    delta_g=[]
    delta_b=[]
    
    prev_r=190
    prev_g=190
    prev_b=190
    
    input_y=870 # daerah yang relatif bersih 

    pixels = image.load()
# create background

    
    for x in range(width):
        output_color=pixels[x,input_y]
        for y in range(height):
            output_image.putpixel((x,y), output_color)
   
    for x in range(width):
        r, g, b = pixels[x, input_y]
        # Process pixel data (e.g., modify or analyze)
        # For example, to set a pixel to black:
        # pixels[x, y] = (0, 0, 0)
        pixels_r.append(r)
        pixels_g.append(g)
        pixels_b.append(b)
    
        delta_r.append(r-prev_r)
        delta_g.append(g-prev_g)
        delta_b.append(b-prev_b)
    
        prev_r = r
        prev_g = g
        prev_b = b

    # draw directly on image
    for x in range(width):
        # red channel
        r=pixels_r[x]
        g=pixels_g[x]
        b=pixels_b[x]
        output_image.putpixel((x, 256-r), (255,0,0))
        output_image.putpixel((x, 512-g), (0,255,0))
        output_image.putpixel((x, 512+256-b), (0,0,255))
    # draw deltas    
        output_image.putpixel((x, 512+256+   40-delta_r[x]*2), (255,0,0))
        draw_g=512+256+80+40-delta_g[x]*2
        if draw_g>1079:
            draw_g=1079
        output_image.putpixel((x, draw_g), (0,255,0))
        output_image.putpixel((x, 512+256+120+40-delta_b[x]*2), (0,0,255))    
    #    image.putpixel((x, 512-g), (0,255,0))
    #    image.putpixel((x, 512+256-b), (0,0,255))

        # draw white line
        image.putpixel((x, y), (255,255,255))
    
    output_image.save(work_output_filename)   
    image.close()
```
