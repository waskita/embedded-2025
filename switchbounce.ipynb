```python
# tutorial step graph:
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/step_demo.html
import matplotlib.pyplot as plt
import numpy as np
```

```python
# bikin sinyal , output ke csv
```

```python
f = open("toggle.csv", "w")
f.write("time,input\n")
for k in range(0,100):
    input=1
    if k<10:
        input=0;
    if k>=20 and k<30:
        input=0;        
    if k>=40 and k<50:
        input=0;        
    if k>=60 and k<70:
        input=0;        
    if k>=80 and k<90:
        input=0;        
    f.write("{},{}\n".format(k,input))
    print(k,input)
f.close()        
```

```python
import pandas as pd
df = pd.read_csv('toggle.csv')
print(df.to_string()) 
```

```python
plt.step(df["time"], df["input"], where='post',color = 'green', label = "input") 
plt.plot(df["time"], df["input"], 'o--',color = 'grey', label = "input", alpha=0.3) 
#plt.step(x, y, where='post', label='post')
plt.savefig("input.jpg")
plt.show()
```
